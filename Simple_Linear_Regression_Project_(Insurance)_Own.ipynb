```python
#data analysis
import pandas as pd
import numpy as np

# machine learning

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# for ploting and visualizing
import matplotlib.pyplot as plt
```

```python
insurance=pd.read_csv("/content/insurance_data.csv")
insurance.head()
```

```python
X=insurance[["Age"]]
y=insurance[["Premium"]]
```

```python
plt.scatter(X,y)
plt.show()
```

```python
# train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
plt.scatter(X_train,y_train,label="Training Data",color="red",alpha=.5)
plt.scatter(X_test,y_test,label="Testing Data",color="blue",alpha=.5)
plt.legend()
plt.title("Train Test Split")
plt.show()
```

```python
# creating Linear Model from training data

model=LinearRegression()

model.fit(X_train,y_train)
```

```python
# visualizing against actual data

y_pred=model.predict(X_test)
```

```python
plt.plot(X_test,y_pred,label="Linear Regression",color="blue")
plt.scatter(X_test,y_test,label="Actual Testing Data",color="green",alpha=.5) # y test
plt.xlabel('Age')
plt.ylabel('Premium')
plt.legend()
plt.grid(True)
plt.show()
 # so have to find the diffrent between y test and y predict ( our model)
```

```python
# evaluating model

mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

print(f"Means Square Error :{mse}")
print(f"Root Mean Square Error :{rmse}")
```

##Conclusion from visualizing

The simple linear regression plot shows the predicted values from your linear regression model (blue line) versus the actual test data (green points).

Observation: The predicted line is fairly close to the actual data points, but there is a noticeable deviation for some points.

Conclusion from visualization: The model captures the general trend of the data, but it is not perfect—some predictions are off by a noticeable margin.
